# Series

The first main data type we will learn about for pandas is the Series data type. Let's import Pandas and explore the Series object.

A Series is very similar to a NumPy array (in fact it is built on top of the NumPy array object). What differentiates the NumPy array from a Series, is that a Series can have axis labels, meaning it can be indexed by a label, instead of just a number location. It also doesn't need to hold numeric data, it can hold any arbitrary Python Object.

Let's explore this concept through some examples:

```python
import numpy as np
import pandas as pd
```

```python
# You can convert a list,numpy array, or dictionary to a Series
labels = ['a','b','c']
# list
my_list = [10, 20, 30]

series = pd.Series(data=my_list)
print(series)

print('\nWith labels/index ')

label_series = pd.Series(data=my_list, index=labels)
print(label_series)
```

```python

dir(pd.Series)[-20:]
```

```python
# array
arr = np.array([10, 20, 30])
# convert to pandas series
pd.Series(arr, labels)
```

```python
# dict
dict_data = {'a':10,'b':20,'c':30}

pd.Series(dict_data)
```

```python
# index 
label_series.index
```

```python
# shape
series.shape
```

```python
# values
series.values
```



# Missing Data

Let's show a few convenient methods to deal with Missing Data in pandas:

```python
df_missing = pd.DataFrame({'A': [1, 2, np.nan],
                           'B': [2, np.nan, None],
                           'C': [1, 2, None]
                          })

df_missing
```

```python
# missing count
df_missing.isna().sum()
```

```python
# remove all missing values in rows
df_missing.dropna()
```

```python
# remove all columns with missing values
df_missing.dropna(axis=1)
```

```python
# remove row with 2 missing values
df_missing.dropna(thresh=2)
```

```python
# replace/fill missing value with 0
# NaN, None, Na = Null

print(df_missing)

new_df = df_missing.fillna(value='Missing')
# new_df.info()
```

```python
new_df
```


# Merging, Joining, and Concatenating

There are 3 main ways of combining DataFrames together: Merging, Joining and Concatenating.

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

- ## Concatenation
Concatenation basically glues together DataFrames. Keep in mind that dimensions should match along the axis you are concatenating on. You can use **pd.concat** and pass in a list of DataFrames to concatenate together:

```python
df1
```

```python
df2
```

```python
df3
```

```python
df_list = [df1, df3, df2]

new_df = pd.concat(df_list)
# new_df.reset_index()
new_df
```

- ## Merging
The **merge** function allows you to merge DataFrames together using a similar logic as merging SQL Tables together. For example:

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K4'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
right
```

```python
left, right
```

```python
# left, right, inner, outer
pd.merge(left, right, how='inner', on='key')
```

- ## Joining
Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame.

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left 
```

```python
right
```

```python
output = left.join(right)
output
```

```python
output.to_csv('example.csv', index=False)
```
